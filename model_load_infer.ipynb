```python
from autogluon.tabular import TabularDataset, TabularPredictor
import random
import os
import numpy as np
import pandas as pd
def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)

seed_everything(9)
```

```python
test_df = pd.read_csv('data/test.csv')
test_df
```

```python
test = TabularDataset(test_df.drop(['SAMPLE_ID'], axis=1))
```

```python
predictor = TabularPredictor.load("saved_model/auto06_saved_model")
pred_y = predictor.predict(test)
```

```python
pred_y = np.where(pred_y < 0, 0, pred_y)
```

```python
print(pred_y)
```

```python
submit = pd.read_csv('./data/sample_submission.csv')
submit['CI_HOUR'] = pred_y
submit.to_csv('out_csvs/auto06.csv', index=False)
```
